- A. Buatlah dua fungsi eksplisit masing-masing memuat perhitungan BMR laki-laki dan BMR Perempuan yang nantinya akan digunakan untuk perhitungan

```python
def BMR_Male (weight,height,age):
  calc = (66 + (137*weight) + (5*height) - (6.8*age))
  return calc
```

```python
BMR_Male(58,162,21)
```

```python
def BMR_Female (weight,height,age):
  calc = (655 + (9.6*weight) + (1.8*height) - (4.7*age))
  return calc
```

```python
BMR_Female (80,188,44)
```
